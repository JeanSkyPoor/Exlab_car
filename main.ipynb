```python
df = pd.read_csv('cars.csv', encoding = "utf-8",delimiter=';' )
df
```

```python
rows = df.shape[0]
print(rows)

for row in range(52, rows):
    item = df.loc[row]
    #print(item)
    type = item[0]
    type_body = item[1]
    mark = item[2]
    fuel = item[3]
    kpp = item[4]
    color = item[5]
    print(type, type_body, mark, fuel, kpp, color)
    
```
